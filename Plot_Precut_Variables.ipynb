```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
MC = np.load('/data/user/xk35/BDT_corrected/post_precut_datasets/MC_GC.npy')
Data = np.load('/data/user/xk35/BDT_corrected/post_precut_datasets/Data_train_GC.npy')
```
